```python
import os
import numpy as np
import tensorflow as tf
import pandas as pd
from scipy.io import wavfile
import scipy.signal as signal

PASTA_AUDIOS = '/home/ma-user/work/audios_wav'
CSV_FILE = '/home/ma-user/work/annotations.csv'
LABEL_ALVO = '5_alert-signal_presence'

print(" ---  TREINAMENTO PESO-PENA ESP32 (ALARMES, SIRENES E BUZINAS) ---")

# 1. MAPEAR ÁUDIOS
mapa_audios = {}
for raiz, _, arquivos in os.walk(PASTA_AUDIOS):
    for f in arquivos:
        if f.endswith('.wav'):
            mapa_audios[f] = os.path.join(raiz, f)

print(f" Áudios físicos mapeados: {len(mapa_audios)}")

# 2. CARREGAR CSV
df = pd.read_csv(CSV_FILE)
df = df[df['split'].isin(['train', 'validate'])]
df_final = (df.groupby('audio_filename')[LABEL_ALVO].mean() >= 0.5).astype(int)

X, y = [], []
contador = 0

def extrair_espectrograma_leve(caminho):
    sr, audio = wavfile.read(caminho)
    audio = audio.astype(np.float32)
    
    if audio.max() > 0:
        audio = audio / np.max(np.abs(audio))
        
    if len(audio.shape) > 1:
        audio = np.mean(audio, axis=1)

    fixo = sr * 2 
    if len(audio) < fixo:
        audio = np.pad(audio, (0, fixo - len(audio)))
    else:
        audio = audio[:fixo]

    f, t, Sxx = signal.spectrogram(audio, fs=sr, nperseg=512, noverlap=256)
    Sxx_db = 10 * np.log10(Sxx + 1e-10)

    Sxx_db = Sxx_db[:48, :] 

    return Sxx_db

# 3. PROCESSAMENTO
print(" Gerando Espectrogramas Otimizados...")
for nome_arquivo, label in df_final.items():
    nome_limpo = os.path.basename(nome_arquivo)

    if nome_limpo in mapa_audios:
        caminho = mapa_audios[nome_limpo]
        try:
            feat = extrair_espectrograma_leve(caminho)
            X.append(feat)
            y.append(label)

            contador += 1
            if contador % 500 == 0:
                print(f"✅ {contador} áudios processados...")
        except Exception as e:
            pass

print(f"\n Total final processado: {contador}")

# 4. TREINAMENTO E CONVERSÃO
if contador > 0:
    X = np.array(X)
    y = np.array(y)
    X = X[..., np.newaxis] 

    print(f"\n Formato da Imagem de Som: {X.shape}")

    model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(8, (3, 3), activation='relu', input_shape=(X.shape[1], X.shape[2], 1)),
        tf.keras.layers.MaxPooling2D((2, 2)),
        
        tf.keras.layers.Conv2D(16, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        
        tf.keras.layers.GlobalAveragePooling2D(), 
        
        tf.keras.layers.Dense(16, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    print(" Iniciando Treinamento...")
    model.fit(X, y, epochs=15, batch_size=32, validation_split=0.2)
    
    print("\n📦 Iniciando Conversão para o ESP32 (TFLite 8-bits)...")
    
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    converter.target_spec.supported_types = [tf.int8]
    
    def representative_dataset():
        for i in range(100):
            yield [np.expand_dims(X[i], axis=0).astype(np.float32)]
            
    converter.representative_dataset = representative_dataset
    tflite_model = converter.convert()

    with open('modelo_touchbike_final_esp32.tflite', 'wb') as f:
        f.write(tflite_model)

    tamanho_kb = len(tflite_model) / 1024
    print("🎉 PRONTO! O cérebro foi gerado!")
    print(f"⭐ TAMANHO FINAL DO MODELO: {tamanho_kb:.2f} KB")

else:
    print("❌ Nenhum áudio processado.")
```
